```python
import pandas as pd

df = pd.read_csv('WRI_CLIMATEWATCH_WIDEF.csv')
df.head()
```

```python
df.columns
```

```python
# Check column names first
print(df.columns)

# Create year column list
year_columns = [str(year) for year in range(1990, 2022)]

# Keep only useful columns
df_clean = df[['REF_AREA', 'REF_AREA_LABEL'] + year_columns]

# Rename columns
df_clean = df_clean.rename(columns={
    'REF_AREA': 'Country_Code',
    'REF_AREA_LABEL': 'Country'
})

# Convert wide format to long format
df_long = df_clean.melt(
    id_vars=['Country_Code', 'Country'],
    var_name='Year',
    value_name='Emissions'
)

# Convert data types
df_long['Year'] = df_long['Year'].astype(int)
df_long['Emissions'] = pd.to_numeric(df_long['Emissions'], errors='coerce')

# Remove missing values
df_long = df_long.dropna()

# Remove duplicates
df_long = df_long.drop_duplicates()

# Sort values
df_long = df_long.sort_values(by=['Country', 'Year'])

# Reset index
df_long = df_long.reset_index(drop=True)

# Preview cleaned data
df_long.head()
```

```python
df_long.to_csv('cleaned_transport_emissions.csv', index=False)
```
